### Machine Leaning Process

- Loading data
- Preprocessing
- Model training
- Model evaluation
- Predictions

```python
# Import needed dependencies
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
# Loading the dataset and displaying the first few rows
iris_data = pd.read_csv('data/raw/iris.csv')
iris_data.head()
```

```python
# Split the data into features (X) and labels (Y)
X = iris_data.drop(columns=['species'])
Y = iris_data['species']
```

```python
# Display the features
X.head()
```

```python
# Display the labels
Y.head()
```

```python
# Initialize the ML model using Logistic Regression
model = LogisticRegression()
```

```python
# Train the model
model.fit(X.values, Y)
```

```python
# Testing prediction with the new model
predictions = model.predict([[4.6, 3.5, 1.5, 0.2]])

# Show the predictions made
print(predictions)
```

```python
# Export prediction model
import joblib

# Save model
joblib.dump(model, "models/prediction.model.pkl")

# Later, load it back
loaded_model = joblib.load("models/prediction.model.pkl")
print(loaded_model.predict([[4.6, 3.5, 1.5, 0.2]]))
```
